```python
#BMI 2.0
height = float(input("Enter your height in meters: "))
weight = float(input("Enter your weight in kgs: "))
bmi = round(weight / height ** 2, 3)
if(bmi < 18.5):
  print("Underweight.")
elif(bmi < 25):
  print("Normal weight.")
elif(bmi < 30):
  print("Slightly overweight.")
elif bmi < 35:
  print("Obese.")
else:
  print("Motachoda.")
```

```python
#difficult challenge: Love calculator
name1 = input("Your name: ")
name2 = input("Your love's name: ")
print("The love calculator is calculating your score...")
name12 = name1.lower() + name2.lower()
# print(name12)
nT = name12.count("t");
nR = name12.count("r");
nU = name12.count("u");
nE = name12.count("e");
dig1 = nT + nR + nU + nE

nL = name12.count("l");
nO = name12.count("o");
nV = name12.count("v");
nE = name12.count("e");
dig2 = nL + nO + nV + nE

score = str(dig1) + str(dig2)
s = int(score)
if s < 10 or s > 90:
  print(f"Your score is {s}, you go like coke and mentos.")
elif s >= 40 and s <= 50:
  print(f"Your score is {s}, you are alright together.")
else:
  print(f"Your score is {s}.")

```
