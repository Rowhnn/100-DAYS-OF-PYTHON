```python
#challenge: how many weeks left if we live upto 90
#ip : 56 ----current age
#op : You have 1768 weeks left

age = int(input())
total_weeks = 90 * 52
age_in_weeks = age * 52
week_left = total_weeks - age_in_weeks
print(f"You have {week_left} weeks left")
```

```python
print("Welcome to the tip calculator.")
bill = float(input("What is the total bill. ₹"))
tip = float(input("How much(%) tip you wanna give? "))
tip = (tip * bill) / 100
people = int(input("How many people to split the bill? "))
bill_per_person = (bill + tip) / people
print(f"Each person should pay: ₹{round(bill_per_person, 2)}.")
```
