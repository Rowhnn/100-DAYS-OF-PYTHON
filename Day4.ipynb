```python
#treasure challege

line1 = ["⬜", "⬜", "⬜"]
line2 = ["⬜", "⬜", "⬜"]
line3 = ["⬜", "⬜", "⬜"]

map = [line1, line2, line3]
print("Hiding your treasure! X marks the spot.")
position = input().upper() #Where do you want to put the treasure?
letters = ['A', 'B', 'C']

letter = position[0]
number = int(position[1])

row = letters.index(letter)
col = number - 1

map[row][col] = "❌"

print(f"{line1}\n{line2}\n{line3}")
```

```python
#Rock Paper Scissors
import random

rock = """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
"""
paper = """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
"""
scissors = """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
"""
# print(scissors)
my_move = int(input())
print("My move: ")
if my_move == 1:
  print(rock)
elif my_move == 2:
  print(paper)
elif my_move == 3:
  print(scissors)

random_move = random.randint(1, 3)
print("Computer's move: ")
if random_move == 1:
  print(rock)
elif random_move == 2:
  print(paper)
elif random_move == 3:
  print(scissors)

if (my_move == 1 and random_move == 3) or (my_move == 2 and random_move == 1) or (my_move == 3 and random_move == 2):
  print("I won.")
elif my_move == random_move:
  print("Draw match.")
else:
  print("I lost.")
```
